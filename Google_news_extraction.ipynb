```
def average(arr):
  a = 0
  for i in range(len(arr)):
    a = a + arr[i]['compound']
  try: 
    a = (a/len(arr))
    a = round(a,1)
    return(a)
  except ZeroDivisionError:
    return(0)
```

```
num = average(arr)
```
